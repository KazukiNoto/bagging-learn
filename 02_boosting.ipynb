# Boosting
- ベースとなる学習器を作成し、その結果をBoostして、次の学習機を作成する手法
- 学習器を順次に学習させる事で、前の学習器よりも精度の高い学習機を構築する手法
- 前の学習器よりも精度をあげるためには、間違えたデータについて再度学習し直す事で精度を上げていく
- 間違えたデータを取得する方法により、いくつかの手法が存在する
    - AdaBoost
    - Gradient Boosting


# Gradient Boosting
- 日本語では勾配ブースティングと言う
- Boostingアルゴリズムを使ってモデルを構築する場合に、残差誤差にフィットするようにモデルを構築します
- 前のモデルが誤分類したデータ点について、より大きな重み付け(勾配を付ける)を行い、次のモデルを計算しなおす

### 残差
- 残差 = 測定値 - 計算値
