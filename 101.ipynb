# 101 - 機械学習のアルゴリズム
- 機械学習のアルゴリズムは多種多様なアルゴリズムがあります
- 値を予測するタイプの回帰アルゴリズムの中だけでも基礎となるアルゴリズムが複数あり、その精度を向上するための手法がさらに複数あります

# アルゴリズムの種類
- Machine Learning
    - Linear Regression(線形回帰)
    - Decision Tree(決定木)
        - Regression Tree(回帰木)
            - Bagging(Booststrap Aggregating)
                - Random Forest
            - Boosting
                - Ada Boosting
                - Gradient Boosting(勾配ブースティング)
                    - XGBoost(ExtreamGradientBoosting)
                    - LightGBM
        - Classificasion Tree(分類木)
    - Neural Network(NN,ニューラルネットワーク)
        - Feedforward Neural Network(順伝播型ニューラルネットワーク)
        - General Regression Neural Network(一般回帰ニューラルネットワーク)
        - Convolutional Neural Network(畳み込みニューラルネットワーク)
        - Recurrent Neural Network(回帰型ニューラルネットワーク)
    - Support Vector Machine(SVM,サポートベクターマシーン)
        - 回帰
            - Support Vector Regression
        - 分類
            - Support Vector Machine
