# Bagging(bootstrap aggregating)
- Baggingとはbootstrap aggregatingの略
- 日本語ではブートストラップ集合と言う
- 近年のMachine Learningのアルゴリズムの中でよく使われる手法の一つ
- 基本となる予測アルゴリズム（決定木など）を組み合わせて、精度をあげるためのアンサンブル学習の手法の一つ

## 流れ
- 学習器に使うためのラーニング用データをBootstrap Samplingによりn個の標本を取り出す
- 取り出したn個の集合から、それぞれ学習器(モデル)を作成し、n個のモデルを構築する
- 作成したn個のモデルを使い予測を行う
- 予測結果の平均を取り、最終結果として算出する

### Bootstrap Sampling
- 母集団となるデータから、母集団から重複を許してランダムにいくつかのデータを取り出して再標本化する手法

### アンサンブル学習
- 複数のモデルを構築して、その結果からデータの予測精度を向上させようとするモデルの場合、その結果をどのように取得するべきかという問題がある
- アンサンブル学習では、その結果を分類の場合は「多数決」、予測の場合は「平均」などで取り出す
